```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


df=pd.read_csv('Books_Data.csv',sep=';')
top_10_books=df.sort_values(by='Book_average_rating', ascending=False).head(10)

plt.figure(figsize=(10,6))
plt.barh(top_10_books['Book Name'],top_10_books['Book_average_rating'],color='skyblue')
plt.xlabel('Average Rating')
plt.ylabel('Book Name')
plt.title('Top 10 Highest Rated Books')
plt.gca().invert_yaxis()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df=pd.read_csv('Books_Data.csv',sep=';')
sales_by_genre=df.groupby('genre')['gross sales'].sum()

plt.figure(figsize=(8,8))
plt.pie(sales_by_genre, labels=sales_by_genre.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Sales Across Book Genres')
plt.axis('equal')
plt.show()
```

```python
fiction_sales=df[df['genre']=='fiction'].groupby('Publishing Year')['gross sales'].sum()

plt.figure(figsize=(10,6))
plt.plot(fiction_sales.index, fiction_sales.values, marker='o',color='green')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Fiction Genre Sales Over the Years')
plt.grid(True)
plt.show
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Book_average_rating'], df['units sold'], color='green')
plt.xlabel('Book Average Rating')
plt.ylabel('Units Sold')
plt.title('Relationship between Book Rating and Sales')
plt.show()
```
